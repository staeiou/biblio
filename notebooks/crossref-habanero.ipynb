```python
import habanero
import pandas as pd
import numpy as np
import json
cr = habanero.Crossref()
```

```python
article = cr.works(ids = '10.1371/journal.pone.0033693')
```

All the top-level data fields:

```python
article['message'].keys()
```

All the data from crossref:

```python
print(json.dumps(article['message'], indent=2))
```

Incoming citations to this article

```python
article['message']['is-referenced-by-count']
```

Papers this article cites

```python
articles_cited_df = pd.DataFrame(article['message']['reference'])
```

```python
articles_cited_df
```

```python
results_list = []
for doi in articles_cited_df['DOI']:
    if doi is not np.nan:
        article_loop = cr.works(ids = doi )
        result = {'doi':doi, 'is-referenced-by-count':article_loop['message']['is-referenced-by-count']}
        results_list.append(result)
        print(result)
```

```python
cited_by_df = pd.DataFrame(results_list)
cited_by_df
```
